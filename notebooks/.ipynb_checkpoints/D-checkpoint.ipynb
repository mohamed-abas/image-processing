#### Histogram and Cumulative Sum functions

```python
# create our own histogram function
def get_histogram(image, bins):
    # array with size of bins, set to zeros
    histogram = np.zeros(bins)
    
    # loop through pixels and sum up counts of pixels
    for pixel in image:
        histogram[pixel] += 1
    
    # return our final result
    return histogram
```

```python
# create our cumulative sum function
def cumsum(a):
    a = iter(a)
    b = [next(a)]
    for i in a:
        b.append(b[-1] + i)
    return np.array(b)
```

```python
def run_image_eqaulization():
    fig = plt.figure(figsize = [7,7])
    gs = plt.GridSpec(2, 3)
    ax1 = fig.add_subplot(gs[0, 0])
    ax2 = fig.add_subplot(gs[0, 1])
    ax3 = fig.add_subplot(gs[0, 2])
    ax4 = fig.add_subplot(gs[1, 0])
    ax5 = fig.add_subplot(gs[1, 1])
    # convert our image into a numpy array
    img_arr = np.asarray(img)

    # put pixels in a 1D array by flattening out img array
    flat = img_arr.flatten()
    ax1.hist(flat, bins=50)
    ax1.set_title('plott flattened image')
    hist = get_histogram(flat, 256)
    ax2.plot(hist)
    ax2.set_title('plott flattened image Histogram')

    # execute the fn
    cs = cumsum(hist)

    # display the result
    ax3.plot(cs)
    ax3.set_title('plott comulative sum of image Histogram')

    nj = (cs - cs.min()) * 255
    N = cs.max() - cs.min()
    # re-normalize the cdf
    cs = nj / N
    cs = cs.astype('uint8')
    ax4.plot(cs)
    ax4.set_title('plott comulative sum of normalized image Histogram')

    # get the value from cumulative sum for every index in flat, and set that as img_new
    img_new = cs[flat]

    # we see a much more evenly distributed histogram
    ax5.hist(img_new, bins=50)
    ax5.set_title('plott equalized image Histogram')

    # put array back into original shape since we flattened it
    img_new = np.reshape(img_new, img_arr.shape)
    # set up side-by-side image display
    fig = plt.figure(figsize = [10,10])
    fig.add_subplot(1,2,1)
    plt.imshow(img[:,:,0], cmap='gray')
    # display the new image
    fig.add_subplot(1,2,2)
    plt.imshow(img_new[:,:,0], cmap='gray')
    plt.show(block=True)
    return img_new[:,:,0]
```

### 1.2 Histogram Matching Functions

```python
def hist_match(source, template):
    """
    Adjust the pixel values of a grayscale image such that its histogram
    matches that of a target image

    Arguments:
    -----------
        source: np.ndarray
            Image to transform; the histogram is computed over the flattened
            array
        template: np.ndarray
            Template image; can have different dimensions to source
    Returns:
    -----------
        matched: np.ndarray
            The transformed output image
    """

    oldshape = source.shape
    source = source.ravel()
    template = template.ravel()

    # get the set of unique pixel values and their corresponding indices and
    # counts
    s_values, bin_idx, s_counts = np.unique(source, return_inverse=True,
                                            return_counts=True)
    t_values, t_counts = np.unique(template, return_counts=True)

    # take the cumsum of the counts and normalize by the number of pixels to
    # get the empirical cumulative distribution functions for the source and
    # template images (maps pixel value --> quantile)
    s_quantiles = cumsum(s_counts).astype(np.float64)
    s_quantiles /= s_quantiles[-1]
    t_quantiles = cumsum(t_counts).astype(np.float64)
    t_quantiles /= t_quantiles[-1]

    # interpolate linearly to find the pixel values in the template image
    # that correspond most closely to the quantiles in the source image
    interp_t_values = np.interp(s_quantiles, t_quantiles, t_values)

    return interp_t_values[bin_idx].reshape(oldshape)
```

```python

def run_hist_match(img_new):
    source = img_new
    source = np.array(source) 
    template = Image.open('images/ref_img.jpg')
    template = np.array(template) 
    matched = hist_match(source, template)
    fig = plt.figure(figsize = [10,10])
    gs = plt.GridSpec(2, 3)
    ax1 = fig.add_subplot(gs[0, 0])
    ax2 = fig.add_subplot(gs[0, 2], sharex=ax1, sharey=ax1)
    ax3 = fig.add_subplot(gs[0, 4], sharex=ax1, sharey=ax1)
    ax1.imshow(source, cmap=plt.cm.gray)
    ax1.set_title('Source')
    ax2.imshow(template, cmap=plt.cm.gray)
    ax2.set_title('template')
    ax3.imshow(matched, cmap=plt.cm.gray)
    ax3.set_title('Matched')
```

####  2.2 log Transformation

```python
def log_trans_img(): 
    img = Image.open('images/im2.jpg')
    img = np.array(img)  

    # Apply log transform. 
    c = 255/(np.log(1 + np.max(img))) 
    log_transformed = c * np.log(1 + img) 

    # Specify the data type. 
    log_transformed = np.array(log_transformed, dtype = np.uint8) 
    # plot images
    f = plt.figure(figsize = [10,10])
    f.add_subplot(1,2, 1)
    plt.imshow(img)
    f.add_subplot(1,2, 2)
    plt.imshow( log_transformed)
    plt.show(block=True)
```

```python
log_trans_img()
```

#### 2.3 Gamma Transformation

```python
def gamma_trans(gamma):    
    # Open the image. 
    img = Image.open('images/im2.jpg')
    img = np.array(img) 
    # Apply gamma correction. 
    corrected = np.array(255*(img / 255) ** gamma, dtype = 'uint8')
    # plot results
    f = plt.figure(figsize = [10,10])
    f.add_subplot(1,2, 1)
    plt.imshow(img)
    f.add_subplot(1,2, 2)
    plt.imshow( corrected)
    plt.show(block=True)
```

```python
gamma_trans(.6)
```

#### 2.4 Piecewise-Linear Transformation

```python
# this function is applied pixel wise
def map_intensity(pixel, r1, s1, r2, s2): 
    if (0 <= pixel and pixel <= r1): 
        return (s1 / r1)*pixel 
    elif (r1 < pixel and pixel <= r2): 
        return ((s2 - s1)/(r2 - r1)) * (pixel - r1) + s1 
    else: 
        return ((255 - s2)/(255 - r2)) * (pixel - r2) + s2 
```

```python
def apply_pwlt(r1, s1, r2, s2):
    # Open the image. 
    img = Image.open('images/im2.jpg')
    img = np.array(img) 
    vec_map_intensity = np.vectorize(map_intensity) 
    corrected = vec_map_intensity(img, r1, s1, r2, s2) 
    # plot results
    f = plt.figure(figsize = [10,10])
    f.add_subplot(1,2, 1)
    plt.imshow(img)
    f.add_subplot(1,2, 2)
    plt.imshow(corrected)
    plt.show(block=True)
```

```python
apply_pwlt(3000,15,10,10)
```

#### 2.5 Bit plane slicing

```python
img =  Image.open('images/im2.jpg')
img = np.array(img)
img = img[:,:,0]
out = []
f = plt.figure(figsize = [10,10])
f.add_subplot(3,3, 1,title = 'Original image')
plt.imshow(img ,  cmap='gray')

for k in range(0, 8):
    # create an image for each k bit plane
    plane = np.full((img.shape[0], img.shape[1]), 2 ** k, np.uint8)
    # execute bitwise and operation
    res = np.bitwise_and(plane, img)
    # multiply ones (bit plane sliced) with 255 just for better visualization
    x = res * 255
    # append to the output list
    out.append(x)
   
for i in range(0, len(out)):
    #imgs = np.hstack(out)
    f.add_subplot(3,3, i+2,title= str(i+1)+ ' bit plane ' )
    plt.imshow(out[i], cmap='gray')
#plt.imshow(imgs)
```

#### 2.6 Intensity level slicing

```python
img =  Image.open('images/im2.jpg')
img = np.array(img)
img = img[:,:,0]
f = plt.figure(figsize = [10,10])
f.add_subplot(1,2, 1)
plt.imshow(img, cmap='gray')
# Find width and height of image
row, column = img.shape
# Create an zeros array to store the sliced image
img1 = np.zeros((row,column),dtype = 'uint8')
 
# Specify the min and max range
min_range = 70
max_range = 120
 
# Loop over the input image and if pixel value lies in desired range set it to 255 otherwise set it to 0.
for i in range(row):
    for j in range(column):
        if img[i,j]>min_range and img[i,j]<max_range:
            img1[i,j] = 255
        else:
            img1[i,j] = 0
            
f.add_subplot(1,2, 2)
plt.imshow(img1, cmap='gray')
```

## 3 Spatial Filters

#### 3.1 Averaging Filter

```python
img =  Image.open('images/im2.jpg')
img = np.array(img)

# Manually implement the mean filter on the grayscale image
gray_avg_img =  img[:,:,0]
gray_img =  img[:,:,0]
for i in range(1, gray_img.shape[0]-1):
    for j in range(1, gray_img.shape[1]-1):
        sum_pix = sum([gray_img[l, k] for l in range(i-1, i+2) for k in range(j-1, j+2)])
        gray_avg_img [i, j] = int(sum_pix/9)
        
rgb_avg_img = img
for i in range(1, img.shape[0]-1):
    for j in range(1, img.shape[1]-1):
        sum_b_pix = sum([img[l, k, 0] for l in range(i-1, i+2) for k in range(j-1, j+2)])
        sum_g_pix = sum([img[l, k, 1] for l in range(i-1, i+2) for k in range(j-1, j+2)])
        sum_r_pix = sum([img[l, k, 2] for l in range(i-1, i+2) for k in range(j-1, j+2)])
        rgb_avg_img [i, j] = [int(sum_b_pix/9), int(sum_g_pix/9), int(sum_r_pix/9)]
        
f = plt.figure(figsize = [10,10])
f.add_subplot(1,3, 1,title = 'Original image')
plt.imshow(img)
f.add_subplot(1,3, 2,title = 'AVG Gray image')
plt.imshow(gray_avg_img, cmap='gray')
f.add_subplot(1,3, 3,title = 'AVG Colored image')
plt.imshow(rgb_avg_img, cmap='gray')
```

#### 3.2 Order-Statistic (Nonlinear) Filters - Median

```python
img =  Image.open('images/im2.jpg')
img = np.array(img)
img = img[:,:,0]
f = plt.figure(figsize = [10,10])
f.add_subplot(1,2, 1,title = 'Original image')
plt.imshow(img ,cmap= 'gray')
m, n = img.shape

img_med = np.zeros([m, n]) 

for i in range(1, m-1): 
    for j in range(1, n-1): 
        temp = [
               img[i-1, j-1], 
               img[i-1, j], 
               img[i-1, j + 1], 
               img[i, j-1], 
               img[i, j], 
               img[i, j + 1], 
               img[i + 1, j-1], 
               img[i + 1, j], 
               img[i + 1, j + 1]] 
          
        temp = sorted(temp) 
        img_med[i, j]= temp[4] 

img_med = img_med.astype(np.uint8) 

f.add_subplot(1,2, 2,title = 'Median Filtered image')
plt.imshow(img_med , cmap= 'gray')
```
