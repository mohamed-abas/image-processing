## Interpolation using nearest Neighbor

```python
import numpy as np
from scipy.interpolate import griddata
import matplotlib.pyplot as plt
from PIL import Image 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def nearestNeighborScaling( source, newWid, newHt ):
    target = Image.new('RGB',(newWid, newHt))
    width, height = source.size
    for x in range(0, newWid):  
        for y in range(0, newHt):
            srcX = int( round( float(x) / float(newWid) * float(width) ) )
            srcY = int( round( float(y) / float(newHt) * float(height) ) )
            srcX = min( srcX, width-1)
            srcY = min( srcY, height-1)
            coordinates = srcX,srcY
            srcColor = source.getpixel(coordinates)
            target.putpixel( (x,y),  srcColor ) 
            #tarPix = getpixel(target, x, y )
            #srcColor = getColor( getPixel(source, srcX, srcY) )
            #setColor( tarPix, srcColor)
    return target
```

```python
ratio = 1.5
source_img = Image.open('images/mona-lisa.jpg')
w,h = source_img.size
new_w = int(w*ratio)
new_h = int(h*ratio)
new_img = nearestNeighborScaling(source_img,new_w,new_h)
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1, 2, 1)
imgplot = plt.imshow(source_img)
ax.set_title('Size  677*1024')
ax = fig.add_subplot(1, 2, 2)
imgplot = plt.imshow(new_img)
ax.set_title('size  1015*1536')
```

## Image transform

```python
name = 'images/mona-lisa.jpg'
img = Image.open(name)  #import image
img = np.array(img.convert('RGB'))

# flip array column 1D
img1 =  img [::-1, ...]

# flip Array rows 1D
img2 =  img1[:, ::-1]
new_img2 = Image.fromarray(img2)
plt.figure(figsize=(10,5))
plt.imshow(new_img2)
```

## Image averaging 

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import scipy.ndimage as ndimage
```

```python
def pxIsInImgRange(x, y):
    if (0<=x) and (x < w): 
            if (0<=y) and (y < h):
                return True
    return False
```

```python
# 1. Load Image
name = 'images/noise.jpg'
img = Image.open(name)  #import image
img = np.array(img) 
h, w = img.shape[:2]
kernel = 25
radius = (kernel-1)/2
img2 = np.zeros((h, w, 3), dtype = 'uint8') #new image to paint on
```

```python
# 2. denoising using average
radius = int(radius)
for x in range (-radius, w+radius):
    for y in range ( -radius, h+radius):
        if pxIsInImgRange(x,y): 
                px = 0
                for vx2 in range (-radius, radius+1):
                    for vy2 in range (-radius, radius+1):
                        x2 = x + vx2
                        y2 = y + vy2
                        if pxIsInImgRange(x2,y2):
                            px = px + (img[y2][x2]/float((kernel*kernel)))
                        else:
                            px = px + 0 
                img2[y][x] = px
```

```python
img  = Image.fromarray(img)
img2 = Image.fromarray(img2)
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1, 2, 1)
imgplot = plt.imshow(img)
ax.set_title('original')

ax = fig.add_subplot(1, 2, 2)
imgplot = plt.imshow(img2)
ax.set_title('25x25')
```
